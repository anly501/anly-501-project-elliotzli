```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df1 = pd.read_csv('../../data/cleaned data/Attempted Suicide 2008-2020 Age.csv')
df2 = pd.read_csv('../../data/cleaned data/Attempted Suicide 2008-2020 Demo.csv')
```

```python
df1 = df1.T
df1.columns = df1.iloc[0]
df1.drop(index=df1.index[0], axis=0, inplace=True)
df2 = df2.T
df2.columns = df2.iloc[0]
df2.drop(index=df2.index[0], axis=0, inplace=True)
```

```python
ax1 = sns.lineplot(data=df1)
ax1.set( xlabel ="Year", ylabel = "Attempted Suicide, in thousands", title ='Attempted Suicide | Age 18-65+')
```

```python
ax2 = sns.lineplot(data=df2)
ax2.set( xlabel ="Year", ylabel = "Attempted Suicide, in thousands", title ='Attempted Suicide | Demographic')
```
