```python
df1 = df[['18-25','26-49','50+']]
df2 = df[['White','Black or African American','Asian','Two or More Races']]
df3 = df[['Northeast Region', 'Midwest Region','South Region', 'West Region']]
df4 = df[['Large Metro Resident','Small Metro Resident', 'Non Metro Resident', 'Urbanized','Less Urbanized']]
```
