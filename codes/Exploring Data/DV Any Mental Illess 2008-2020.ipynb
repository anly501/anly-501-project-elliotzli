```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../../data/cleaned data/Any Mental Illness 2008-2020.csv')
```

```python
df
df = df.set_index(df.columns[0])
df1 = df[0:5]
df1 = df1.T

fig, ax = plt.subplots(figsize=(8, 5))
fig = sns.lineplot(data=df1,palette='ch:s=.25,rot=-.25')
fig.set( xlabel ="Year", ylabel = "Population with Any Mental Illness, in thousands", title ='Population with Any Mental Illness | Demographic')

df2 = df[5:11]
df2 = df2.T

fig, ax = plt.subplots(figsize=(8, 5))
fig = sns.lineplot(data=df2,palette='ch:s=.25,rot=-.25')
fig.set( xlabel ="Year", ylabel = "Population with Any Mental Illness, in thousands", title ='Population with Any Mental Illness | Race & Ethnicity')

df3 = df[11:15]
df3 = df3.T

fig, ax = plt.subplots(figsize=(8, 5))
fig = sns.lineplot(data=df3,palette='ch:s=.25,rot=-.25')
fig.set( xlabel ="Year", ylabel = "Population with Any Mental Illness, in thousands", title ='Population with Any Mental Illness | Geographic Location')

df4 = df[15:20]
df4 = df4.T

fig, ax = plt.subplots(figsize=(8, 5))
fig = sns.lineplot(data=df4,palette='ch:s=.25,rot=-.25')
fig.set( xlabel ="Year", ylabel = "Population with Any Mental Illness, in thousands", title ='Population with Any Mental Illness | Urbanization Level')

df5 = df[20:24]
df5 = df5.T

fig, ax = plt.subplots(figsize=(8, 5))
fig = sns.lineplot(data=df5,palette='ch:s=.25,rot=-.25')
fig.set( xlabel ="Year", ylabel = "Population with Any Mental Illness, in thousands", title ='Population with Any Mental Illness | Educational Level')
```
